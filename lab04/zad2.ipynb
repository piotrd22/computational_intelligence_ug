```python
from matplotlib import pyplot as plt
import random
from aco import AntColony
import time
```

```python
plt.style.use("dark_background")


COORDS = (
    (20, 52),
    (43, 50),
    (20, 84),
    (70, 65),
    (29, 90),
    (87, 83),
    (73, 23),
    (53, 85),
    (14, 43),
    (8, 29),
    (25, 13)
)
```

```python
def random_coord():
    r = random.randint(0, len(COORDS))
    return r


def plot_nodes(w=12, h=8):
    for x, y in COORDS:
        plt.plot(x, y, "g.", markersize=15)
    plt.axis("off")
    fig = plt.gcf()
    fig.set_size_inches([w, h])


def plot_all_edges():
    paths = ((a, b) for a in COORDS for b in COORDS)

    for a, b in paths:
        plt.plot((a[0], b[0]), (a[1], b[1]))
```

```python
plot_nodes()

# colony = AntColony(COORDS, ant_count=300, alpha=0.5, beta=1.2,
#                    pheromone_evaporation_rate=0.40, pheromone_constant=1000.0,
#                    iterations=300)

colony_args = [[300, 0.3, 1, 0.2, 700, 300], [300, 0.5, 1.4, 0.6, 1100, 300], 
               [300, 0.4, 1.2, 0.4, 900, 300], [300, 0.5, 1.2, 0.4, 1000, 300]]

times=[]
for i in colony_args:
    time_start = time.time()
    ant_count, alpha, beta, pheromone_evaporation_rate, pheromone_constant, iterations = i
    colony = AntColony(COORDS, 
                       ant_count=ant_count, 
                       alpha=alpha, 
                       beta=beta,
                       pheromone_evaporation_rate=pheromone_evaporation_rate, 
                       pheromone_constant=pheromone_constant, 
                       iterations=iterations)
    times.append(time.time() - time_start)
```

```python
print(times)
```

Po kilku próbach wygenewrowanych z taką samą ilością mrówek oraz ilością iteracji najlepiej wypadł zestaw 3, jednak szczerze po kilku próbach często szybsze i lepsze wyniki dawała próba nr 1, więc nie jest powiedziane, że zawsze te same próbki wygenerują ten sam wynik i czas.

```python
optimal_nodes = colony.get_path()

for i in range(len(optimal_nodes) - 1):
    plt.plot(
        (optimal_nodes[i][0], optimal_nodes[i + 1][0]),
        (optimal_nodes[i][1], optimal_nodes[i + 1][1]),
    )


plt.show()
```
